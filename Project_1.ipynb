#1. Movies Vs TV Shows on Netflix


```python
#improting liabrary
import matplotlib.pyplot as plt
import seaborn as sns

# Count Movies and TV Shows
content_count = df['type'].value_counts()

# Create Bar Chart
plt.figure(figsize=(6,4))
sns.countplot(data=df,x='type',hue='type',palette='Set2')

# Set Labels
plt.title('Movies vs TV Shows on Netflix',color="red",size=15)
plt.xlabel('Content Type',color='green',size=12)
plt.ylabel('Number of Titles',color='green',size=12)

plt.grid()

plt.show()
```

This Chart compare the number of Movies and TV Shows available on Netflix. We used a Seaborn Count plot to count the occurrences of each content type. From the chart, we can clearly see that Movies have a much higher count than TV Shows.
"Approximately 70% of the content on Netflix consists of Movies, while around 30% consists of TV Shows."

#2.Top Countries Producing Content

```python
from matplotlib import figure
# Count the frequency of each country
# value_counts() counts how many times each country appears
# head(10) selects the top 10 countries
df['country'].value_counts().head(10).plot(kind='bar',color='teal',figsize=(8,6))

# Set labels and title

# Add a title to the chart
plt.title('Top 10 Countries',color='red',size=15)

# Add labels to the axes
plt.xlabel('Country',color='green',size=12)
plt.ylabel('Number of Titles',color='green',size=12)

#Display the chart
plt.show()
```

This bar chart shows the top 10 countries producing content available on Netflix. The x-axis represents countries, and the y-axis represents the number of Movies and TV Shows. "From the analysis, we can see that the United States is the largest content producer with approximately 2,800 titles. "India ranks second with around 1,000 titles, demonstrating its strong presence on the Netflix platform."
"The 'Unknown' category represents records where country information is not available in the dataset."
"Other countries such as the United Kingdom, Japan, South Korea, Canada, Spain, France, and Mexico also contribute significant content, but their numbers are comparatively lower."

#3.Content growth over years?

```python
import plotly.express as px

# Count content by release year
content_growth = (
    df['release_year']
    .value_counts()
    .sort_index()
    .reset_index()
)

content_growth.columns = ['Release Year', 'Number of Titles']

# Interactive Line Chart
fig = px.line(
    content_growth,
    x='Release Year',
    y='Number of Titles',
    title='Content Growth Over the Years',
    markers=True
)

fig.update_layout(
    xaxis_title='Release Year',
    yaxis_title='Number of Titles',
    template='plotly_white',
    width=900,
    height=600
)

fig.show()
```

"This line chart represents the number of Movies and TV Shows released each year on Netflix. The x-axis shows the release year, and the y-axis shows the number of titles."
"From the chart, we can observe that content releases remained relatively low from the 1920s to the early 2000s."
"After 2010, there is a significant increase in the number of releases, indicating rapid expansion of Netflix's content library."
"The highest number of releases occurs around 2018–2019, where more than 1,100 titles were added.""After 2019, the number of releases shows a slight decline, but Netflix continues to maintain a large volume of content."

#4. Most Popular Content Categories (Genres)?

```python
import pandas as pd
import plotly.express as px

# Split multiple categories into separate rows
categories = df['listed_in'].str.split(', ').explode()

# Count top 10 categories
top_categories = (
    categories.value_counts()
    .head(10)
    .reset_index()
)

top_categories.columns = ['Content Category', 'Number of Titles']

# Plotly Horizontal Bar Chart
fig = px.bar(
    top_categories,
    x='Number of Titles',
    y='Content Category',
    orientation='h',
    color='Number of Titles',
    color_continuous_scale='prgn',
    title='Top 10 Most Popular Content Categories on Netflix'
)

fig.update_layout(
    xaxis_title='Number of Titles',
    yaxis_title='Content Category',
    width=900,
    height=600,
    yaxis=dict(autorange='reversed')
)

fig.show()
```

"This chart shows the top 10 content categories available on Netflix. International Movies are the most dominant category with nearly 2,750 titles, followed by Dramas and Comedies. The results indicate that Netflix heavily invests in globally accessible content and storytelling-focused genres. Additionally, the strong presence of International TV Shows reflects the growing popularity of regional and non-English content. These insights help Netflix understand audience preferences and guide future content acquisition and production strategies."

#5. Rating Distribution?

```python
#import plotly library
import plotly.express as px

# value_counts() counts how many times each country appears
rating_count = df['rating'].value_counts().reset_index()

#rating count for the column count
rating_count.columns = ['Rating', 'Count']

#adding figure, title, values
fig = px.pie(
    rating_count,
    names='Rating',
    values='Count',
    title='Netflix Rating Distribution'
)

 #show the figure
fig.show()
```

"This pie chart shows the distribution of Netflix content ratings. The largest segment is TV-MA, accounting for approximately 36% of all titles, indicating a strong focus on mature audiences. TV-14 is the second-largest category at around 25%, highlighting Netflix's appeal among teenagers and young adults. Family-oriented ratings such as TV-PG, PG, and TV-Y represent a smaller portion of the library. Overall, the chart suggests that Netflix's content strategy is primarily designed for adult and young adult viewers, while also maintaining a smaller collection of family-friendly content."

#6.Top 10 genres on Netflix ?

```python
import plotly.express as px

# Split genres and count frequency
genres = df['listed_in'].str.split(', ').explode()

top_genres = genres.value_counts().head(10).reset_index()
top_genres.columns = ['Genre', 'Count']

# Plotly Bar Chart
fig = px.bar(
    top_genres,
    x='Genre',
    y='Count',
    title='Top 10 Genres on Netflix',
    labels={
        'Genre': 'Genre',
        'Count': 'Number of Titles'
    },
    color='Count',
    hover_data=['Count'],
    text='Count'
)

# Format text shown on bars
fig.update_traces(
    texttemplate='%{text}',
    textposition='outside'
)

# Improve layout
fig.update_layout(
    xaxis_title='Genre',
    yaxis_title='Number of Titles',
    xaxis_tickangle=-45
)

fig.show()
```

"This chart highlights the top 10 genres available on Netflix. International Movies lead with 2,752 titles, followed by Dramas with 2,427 titles and Comedies with 1,674 titles. The results show that Netflix heavily emphasizes globally appealing content, particularly international productions and storytelling-focused genres. The strong presence of International TV Shows also reflects the growing popularity of regional content among viewers worldwide. These insights help Netflix make strategic decisions regarding content acquisition, production, and audience targeting."

#7. Content Rating Vs Content Type

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a pivot table
heatmap_data = pd.crosstab(df['rating'], df['type'])

# Plot Heatmap
plt.figure(figsize=(10, 8))

sns.heatmap(
    heatmap_data,
    annot=True,
    cmap='YlOrRd',
    fmt='d'
)

plt.title('Netflix Content Rating vs Content Type')
plt.xlabel('Content Type')
plt.ylabel('Rating')

plt.show()
```

"This heatmap compares Netflix content ratings across Movies and TV Shows. The darkest areas represent the highest concentration of titles. We can see that TV-MA is the dominant rating category, with over 2,000 movies and more than 1,100 TV shows. TV-14 is the second-largest category, indicating a strong focus on teenagers and young adults. Movies consistently outnumber TV shows across most rating categories. Family-friendly ratings such as TV-Y and TV-G represent a smaller share of the content library. Overall, the analysis shows that Netflix's content strategy is primarily centered on mature and young adult audiences."

#8. Distribution of Release Years

```python
#distribution of release years with histogram
plt.figure(figsize=(10,6))

#histogram with bins and KDE map
sns.histplot(df['release_year'], bins=20, kde=True)

#add label and title
plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')

#Display the chart
plt.show()
```

"This histogram illustrates the distribution of Netflix titles by release year. We can see that the majority of content was released after 2010, with a particularly strong concentration between 2018 and 2020. Very few titles originate from earlier decades, indicating that Netflix's library is primarily composed of modern content. This trend suggests that Netflix prioritizes recent releases and continuously updates its catalog to align with current viewer preferences and market demand."
